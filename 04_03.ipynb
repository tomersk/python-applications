```python
import numpy 
import matplotlib.pyplot 
from pyflo import system
from pyflo.nrcs import hydrology

runoff = numpy.array([
    (0.00, 0.000),
    (0.1, 0.03),
    (0.2, 0.1),
    (0.3, 0.19),
    (0.4, 0.31),
    (0.5, 0.47),
    (0.6, 0.66),
    (0.7, 0.82),
    (0.8, 0.93),
    (0.9, 0.99),
    (1, 1),
    (1.1, 0.99),
    (1.2, 0.93),
    (1.3, 0.86),
    (1.4, 0.78),
    (1.5, 0.68),
    (1.6, 0.56),
    (1.7, 0.46),
    (1.8, 0.39),
    (1.9, 0.33),
    (2, 0.28),
    (2.2, 0.207),
    (2.4, 0.147),
    (2.6, 0.107),
    (2.8, 0.077),
    (3, 0.055),
    (3.2, 0.04),
    (3.4, 0.029),
    (3.6, 0.021),
    (3.8, 0.015),
    (4, 0.011),
    (4.5, 0.005),
    (5, 0)
])
basin = hydrology.Basin(
    area=4.6,
    cn=85.0,
    tc=2.3,
    runoff_dist=runoff,
    peak_factor=484.0
)

unit_hydrograph = basin.unit_hydrograph(interval=0.3)
from matplotlib import pyplot

x = unit_hydrograph[:, 0]
y = unit_hydrograph[:, 1]
pyplot.plot(x, y, 'k')
pyplot.plot(x, y, 'bo')
pyplot.title(r'Unit Hydrograph')
pyplot.xlabel(r'Time ($hr$)')
pyplot.ylabel(r'Discharge ($\frac{ft^{3}}{s}$)')
pyplot.show()
```
